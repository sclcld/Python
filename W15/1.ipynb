```python
"""
    Il dataset insurance.csv dei beginner_datasets contiene dati su caratteristiche e abitudini delle persone rispetto ai costi individuali per le cure
    mediche come premio per le assicurazioni sulla salute. 
    • Visualizziamone le dimensioni, un'anteprima, e osserviamo i nomi di colonna; 
    • Quali sono le medie di charges rispetto a region? Ci sono differenze significative? 
    • E rispetto a smoker? E a sex? 
    • Quali sono i descrittori statistici di bmi? 
"""    



df = pd.read_csv('insurance.csv')  
print(df.shape)
print(df.sample())
print("Nomi delle colonne:", df.columns)
print("Medie charges groupby region",df.groupby('region')['charges'].mean())
print("Medie di charges groupby smoker:",df.groupby('smoker')['charges'].mean() )
medie_per_sex = df.groupby('sex')['charges'].mean()
print("Medie di charges groupby sex:", df.groupby('sex')['charges'].mean())
desc_bmi = df['bmi'].describe()
print(f"Descrittori bmi: {desc_bmi}")
```

```python
df = pd.read_csv('iris.csv')  
print(df.shape)
print(df.sample(),"\n")
"""
    • La media della lunghezza dei petali di tutto il dataset 
    • La media della lunghezza dei petali per ogni specie di Iris, utilizzando il metodo .groupby() 
    • Media, minimo e massimo della larghezza dei sepali per ogni specie, utilizzando .groupby() e .agg()
"""    
print(f"Media lunghezza petali : {df['petal_length'].mean():.2f}", )
print("\nMedia della lunghezza petali per specie:\n", df.groupby('species')['petal_length'].mean())
print("\nStatistiche larghezza sepali per ogni specie di Iris: \n",df.groupby('species')['sepal_width'].agg(['mean', 'min', 'max']))

```

```python
"""
    Nei beginner_datasets, il dataset wine.csv contiene delle analisi organolettiche su diverse qualità di vini: 
    • Qual è la media di concentrazione alcolica per ogni qualità? Ci sono differenze? E rispetto alla media totale? •
    C'è differenza nella concentrazione alcolica per vini bianchi e vini rossi? • Rifacendo le analisi delle domande precedenti ma per il pH, cambia qualcosa? 
    • E per i solfati?
"""

df = pd.read_csv('wine.csv')
print(df.head())
print("\nMedia concentrazione alcolica per qualità:\n",df.groupby('quality')['alcohol'].mean())
print("\nMedia concentrazione alcolica:", df['alcohol'].mean())
print(f"Media concentrazione alcolica per bianchi e rossi: \n {df.groupby('type')['alcohol'].mean()}")
print("\nMedia pH per qualità:\n", df.groupby('quality')['pH'].mean())
print("\nMedia pH:\n", df['pH'].mean())
print("\nMedia di solfati per qualità:\n", df.groupby('quality')['sulphates'].mean())
print("\nMedia totale di solfati:\n", df['sulphates'].mean())
```

```python
employees_df = pd.DataFrame({
    'employee_id': [101, 102, 103, 104, 105],
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Emma'],
    'department_id': [1, 2, 1, 2, 3]
})

departments_df = pd.DataFrame({
    'department_id': [1, 2, 3],
    'department_name': ['HR', 'IT', 'Finance'],
    'location': ['New York', 'San Francisco', 'Chicago']
})
"""
    • Unire questi DataFrame in base alla colonna comune department_id, in modo da avere nel risultato informazioni sia 
      sui dipendenti che sui dipartimenti, usando la funzione .merge()
"""
print( pd.merge(employees_df, departments_df, on='department_id'))
```

```python
""" 
    • Per ogni DataFrame, trasformare la colonna department_id nell'indice, facendo in modo che la modifica sia permanente; 
    poi unire i due dataset mediante il metodo .join()
"""

employees_df.set_index('department_id', inplace=True)
departments_df.set_index('department_id', inplace=True)

print(employees_df.join(departments_df, how='inner'))
```

```python
"""
    • Dal database AdventureWorksDW importiamo le tabelle dimemployee e dimemployeesalesterritory come DataFrame 
    • Effettuiamo un join tra i due DataFrame usando le colonne EmployeeKey 
    • Controlliamo la dimensione del DataFrame risultante: è quella attesa? 
    • Importiamo ora la tabella dimsalesterritory ed effettuiamo un join tra questa e il DataFrame risultante della join precedente, usando le colonne SalesTerritoryKey 
"""


import dotenv
import os
from sqlalchemy import create_engine

dotenv.load_dotenv(override= True)

username = os.getenv("username")
password = os.getenv("password")
host = os.getenv("host")
dbname = os.getenv("dbname")

conn_string = f"mysql+pymysql://{username}:{password}@{host}/{dbname}"
db_engine = create_engine(conn_string)
query = 'SELECT * FROM dimemployee'
query2 = 'SELECT * FROM dimemployeesalesterritory'
query3= 'SELECT * FROM dimsalesterritory'

dimemployee = pd.read_sql(query,db_engine)
dimemployeesalesterritory = pd.read_sql(query2, db_engine)
dimsalesterritory = pd.read_sql(query3, db_engine)

db_engine.dispose()
```

```python
merged = pd.merge(dimemployee, dimemployeesalesterritory, on='EmployeeKey')
print(merged.shape)
```

```python
merged = pd.merge(merged,dimsalesterritory, on="SalesTerritoryKey")
```

```python
print(merged.shape)
```

```python
"""
    • Su questo DataFrame contare quanti dipendenti ci sono per ogni paese (country) e per ogni regione (region)
"""
print(merged.groupby(['SalesTerritoryCountry', 'SalesTerritoryRegion']).size())
```
